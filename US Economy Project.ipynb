### Define a function that makes a Dashboard

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show, output_notebook
output_notebook()
```

```python
def make_dashboard(x, gdp_change, unemployment, title, filename):
    output_file(filename)
    p = figure(title=title, x_axis_label='year', y_axis_label= '%')
    p.line(x.squeeze(), gdp_change.squeeze(), color='firebrick', line_width=4, legend = '% GDP changes')
    p.line(x.squeeze(), unemployment.squeeze(), line_width=4, legend = '% unemployment')
    show(p)
```
